## ** Dimensional Modelling (Star Schama)** 

```python
%sql
-- View: Customer
CREATE OR REPLACE VIEW pysparkdbt.gold.customers AS 
SELECT 
  customer_id,
  full_name AS customer_name,
  email,
  domain,
  phone_number,
  city,
  signup_date,
  last_updated_timestamp,
  process_timestamp AS dwh_process_timestamp
FROM pysparkdbt.silver.customers
```

```python
%sql
-- View: drivers
CREATE OR REPLACE VIEW pysparkdbt.gold.drivers AS
SELECT 
  driver_id,
  full_name,
  phone_number,
  city,
  vehicle_id,
  driver_rating,
  last_updated_timestamp,
  process_timestamp AS dwh_process_timestamp
FROM pysparkdbt.silver.drivers
```

```python
%sql
-- View: locations
CREATE OR REPLACE VIEW pysparkdbt.gold.locations AS
SELECT 
  location_id,
  city,
  state,
  country,
  latitude,
  longitude,
  last_updated_timestamp,
  process_timestamp AS dwh_process_timestamp
FROM pysparkdbt.silver.locations
```

```python
%sql
-- View: payments
CREATE OR REPLACE VIEW pysparkdbt.gold.payments AS 
SELECT
  payment_id,
  trip_id,
  customer_id,
  payment_method,
  payment_status,
  amount,
  online_payment_status,
  transaction_time,
  last_updated_timestamp,
  process_timestamp AS dwh_process_timestamp
FROM pysparkdbt.silver.payments
```

```python
%sql
-- View: vehicles
CREATE OR REPLACE VIEW pysparkdbt.gold.vehicles AS
SELECT
  vehicle_id,
  license_plate,
  model AS vehicle_model,
  make AS manufracturur,
  year,
  vehicle_type,
  last_updated_timestamp,
  process_timestamp AS dwh_process_timestamp
FROM pysparkdbt.silver.vehicles
```

```python
%sql
-- View: trips
CREATE OR REPLACE VIEW pysparkdbt.gold.trips AS 
SELECT
  trip_id,
  driver_id,
  customer_id,
  vehicle_id,
  start_location,
  end_location,
  distance_km,
  fare_amount,
  trip_start_time,
  trip_end_time,
  last_updated_timestamp,
  process_timestamp AS dwh_processs_timestamp
FROM pysparkdbt.silver.trips
```
